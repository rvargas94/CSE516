### Implementing KNN algorithm
Implementing KNN with k = 3, 5, 9, 13

```python
#eDistance takes two points and returns the euclidian distance
def eDistance(initial_point,test_point):
    a = initial_point
    b = test_point
    dist = np.linalg.norm(a-b) 
    #print a,"\n"
    #print b

    return dist
    
testDistance = eDistance(trainData3D.iloc[1],testData3D.iloc[2])
print testDistance
```

```python
def nearestNeighbors(tSet,testInstance,k):
    distances = []
    for i in range(len(tSet)):
        iDistance = eDistance(tSet.iloc[i],testInstance)
        distances.append((tSet.iloc[i],iDistance))
    
    distances.sort(key=operator.itemgetter(1))
    neighbors = []
    for i in range(k):
        neighbors.append(distances[i][0])
        distances[i]
    return neighbors

testNeighbor = pd.DataFrame(nearestNeighbors(trainData3D,testData3D.iloc[99],9))
testNeighbor.head(9)
```

```python
def prediction(neighbors):
    votes = {}
    for i in range(len(neighbors)):
        label = neighbors.index.values[i]
        if label in votes:
            votes[label] += 1
        else:
            votes[label]  = 1
    votes_sorted = sorted(votes.iteritems(), key=operator.itemgetter(1),reverse=True)
    return votes_sorted[0][0]

testPrediction = prediction(testNeighbor)
print testPrediction
```

```python
def accuracy(testSet,predictions):
    correct_label_rate = 0
    for i in range(len(testSet)):
        #print i,testSet.values[i][0], predictions.iloc[i][0]
        if testSet.values[i][0] == predictions.iloc[i][0]:
            correct_label_rate += 1
            #print correct_label_rate
    correct_label_rate /= float(len(testSet))
    return correct_label_rate*100
```

```python

def knnAlgo(trSet,teSet,k):
    label_predictions = []
    print "First 20 predictions of k="+str(k)+" ..."
    for x in range(len(teSet)):
        kNeighbors = pd.DataFrame(nearestNeighbors(trSet,teSet.iloc[x],k))
        current_label = prediction(kNeighbors)
        label_predictions.append(current_label)
        if x < 20:
            print "Prediction is " + current_label + " Actual is " + testTarget.iloc[x].ravel()
    label_predictions = pd.DataFrame(label_predictions)
    return accuracy(testTarget,label_predictions) 

knn3 = knnAlgo(trainData3D,testData3D,3)
print knn3, "\n"

knn5 = knnAlgo(trainData3D,testData3D,5)
print knn5, "\n"

knn9 = knnAlgo(trainData3D,testData3D,3)
print knn9, "\n"

knn13 = knnAlgo(trainData3D,testData3D,3)
print knn13, "\n"
```
